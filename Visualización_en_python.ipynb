```python
import matplotlib.pyplot as plt
import numpy as np
x=np.array([1,2,3])
y=np.array([2,30,4])
plt.plot(x,y)
plt.show()
```

```python
plt.plot(x,y,color='red',marker='o',markerfacecolor='green',markersize=15)
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.title('Graph')
plt.grid('true')
plt.show()
```

```python
y=np.array([35,25,25,15])
mylabels=['Apples','Bananas','Cherries','Dates']

plt.pie(y,labels=mylabels)
plt.show()
```

```python
plt.pie(y,labels=mylabels,startangle=90)
plt.show()
```

```python
x=np.random.normal(170,10,50)
plt.hist(x)
plt.show()
```

```python
x=np.array(['A','B','C','D'])
y=np.array([10,8,2,15])
plt.bar(x,y,color='red')
plt.show()
```

```python
plt.barh(x,y,color='red')
plt.show()
```
